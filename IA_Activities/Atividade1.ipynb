## O primeiro bloco de código será para o mapeamento e preparação dos dados de currículos de pessoas com Nome, Idade, Experiência, Email e Sexo.

## Os gráficos usam os dados para demonstrar o nível de educação, idade e ano de entrada dos funcionários.

```python
# Importação de Bibliotecas 

import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt


# Leitura de Dataset

df = pd.read_csv("curriculos.csv",sep=",")  

 
# Data Understanding - Compreensão dos Dados 

df.head() 
df.info() 
df.describe() 
df.describe(include='O') 

print(df['Nome'].value_counts()) 
print(df['Sexo'].value_counts()) 


# Data Preparation

df.isnull().sum() # para totalizar os valores nulos em cada feature (variável) 
df[df.duplicated(keep='first')] # para identificar linhas duplicadas 

# Como serão dados de pessoas, os nomes, idades, experiências, etc... podem se repetir, portanto
# Não irei considerar remover as linhas duplicadas
```

## O bloco a seguir mostra a distribuição de idade dentro da lista de candidatos.

```python
# Data Visualization 


# Distribuição da Variável Education Level 

plt.figure(figsize=(12, 6))  # Define o tamanho da figura
sns.histplot(data=df, x="Idade", bins=10, kde=True, color="skyblue")
plt.title("Distribuição de Idade")
plt.xlabel("Idade")
plt.ylabel("Contagem")
plt.show()
```

```python
# Função para verificar se uma linha possui a profissão desejada
def tem_profissao(row, profissao):
    return profissao in row['Experiência']

# Filtrar o DataFrame para contar quantas pessoas são Engenheiras ou Engenheiros
engenheiros = df[df.apply(lambda row: tem_profissao(row, 'Engenheiro'), axis=1)]

# Filtrar o DataFrame para contar quantas pessoas são Desenvolvedoras ou Desenvolvedores
desenvolvedores = df[df.apply(lambda row: tem_profissao(row, 'Desenvolvedor'), axis=1)]

# Criar um gráfico de barras para mostrar o número de Engenheiros e Desenvolvedores
categorias = ['Engenheiros', 'Desenvolvedores']
contagens = [len(engenheiros), len(desenvolvedores)]

plt.figure(figsize=(10, 6))
sns.barplot(x=categorias, y=contagens, palette="Set2")
plt.title("Número de Engenheiros e Desenvolvedores (Masculino)")
plt.ylabel("Contagem")
plt.show()
```

```python
# Função para verificar se uma linha possui a profissão desejada
def tem_profissao(row, profissao):
    return profissao in row['Experiência']

# Filtrar o DataFrame para contar quantas pessoas são Engenheiras ou Engenheiras
engenheiras = df[df.apply(lambda row: tem_profissao(row, 'Engenheira'), axis=1)]

# Filtrar o DataFrame para contar quantas pessoas são Desenvolvedoraas ou Desenvolvedoraes
desenvolvedoraes = df[df.apply(lambda row: tem_profissao(row, 'Desenvolvedora'), axis=1)]

# Criar um gráfico de barras para mostrar o número de Engenheiras e Desenvolvedoraes
categorias = ['Engenheiras', 'Desenvolvedoraes']
contagens = [len(engenheiras), len(desenvolvedoraes)]

plt.figure(figsize=(10, 6))
sns.barplot(x=categorias, y=contagens, palette="Set2")
plt.title("Número de Engenheiras e Desenvolvedoraes (Masculino)")
plt.ylabel("Contagem")
plt.show()
```

```python
# Função para verificar se uma linha possui a profissão desejada
def tem_profissao(row, profissao):
    return profissao.lower() in row['Experiência'].lower()

# Filtrar o DataFrame para contar quantas pessoas são Engenheiras ou Engenheiros
engenheiras = df[df.apply(lambda row: tem_profissao(row, 'engenheira'), axis=1)]
engenheiros = df[df.apply(lambda row: tem_profissao(row, 'engenheiro'), axis=1)]

# Filtrar o DataFrame para contar quantas pessoas são Desenvolvedoras ou Desenvolvedores
desenvolvedoras = df[df.apply(lambda row: tem_profissao(row, 'desenvolvedora'), axis=1)]
desenvolvedores = df[df.apply(lambda row: tem_profissao(row, 'desenvolvedor'), axis=1)]

# Contar o número de Engenheiras e Engenheiros
contagem_engenheiras = len(engenheiras)
contagem_engenheiros = len(engenheiros)

# Contar o número de Desenvolvedoras e Desenvolvedores
contagem_desenvolvedoras = len(desenvolvedoras)
contagem_desenvolvedores = len(desenvolvedores)

# Criar um gráfico de barras para mostrar a comparação entre sexo feminino e masculino
categorias = ['Engenheiros', 'Desenvolvedores']
contagens_feminino = [contagem_engenheiras, contagem_desenvolvedoras]
contagens_masculino = [contagem_engenheiros, contagem_desenvolvedores]

bar_width = 0.35
index = range(len(categorias))

plt.figure(figsize=(12, 6))
bar1 = plt.bar(index, contagens_feminino, bar_width, label='Feminino', color='lightcoral')
bar2 = plt.bar(index, contagens_masculino, bar_width, label='Masculino', color='lightblue', alpha=0.7, align='edge')

plt.xlabel("Profissão")
plt.ylabel("Contagem")
plt.title("Comparação entre Sexo Feminino e Masculino - Engenheiros e Desenvolvedores")
plt.xticks(index, categorias)
plt.legend()
plt.show()
```
